```python

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import metrics
import tensorflow as tf
from tensorflow import keras

df = pd.read_excel("C:/Users/7kingdom/Downloads/Raisin_Dataset/Raisin_Dataset.xlsx")
df
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

label=LabelEncoder()
df['Class']=label.fit_transform(df['Class'])

df
```

```python
x = df
x = x.drop(columns=['Class'], axis=1)

x
```

```python
y = df['Class']

y
```

```python
X = pd.get_dummies(x).values

X
```

```python
Y = pd.get_dummies(y).values

Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=0) 
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(X_train.shape[1],)),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(Y.shape[1], activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

model.summary()
```

```python
history = model.fit(X_train, y_train, epochs=50, validation_data=(X_test, y_test))
```

```python
test_loss, test_acc = model.evaluate(X_test, y_test)
print("Test accuracy:", test_acc)
```

```python
from sklearn.metrics import f1_score

y_pred = model.predict(X_test)
y_pred = np.argmax(y_pred, axis=1)
y_true = np.argmax(y_test, axis=1)

f1 = f1_score(y_true, y_pred, average='weighted')

print("F1 Score:", f1)
```

```python
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)

y_true = np.argmax(y_test, axis=1)

f1 = f1_score(y_true, y_pred_classes)

print("F1 Score:", f1)
```

```python
y_pred_prob = model.predict(X_test)

best_f1 = 0.0
best_threshold = 0.0

for threshold in np.arange(0.1, 1.0, 0.1):
    y_pred_classes = (y_pred_prob[:, 1] > threshold).astype(int)
    f1 = f1_score(y_true, y_pred_classes)
    
    if f1 > best_f1:
        best_f1 = f1
        best_threshold = threshold

print("Best F1 Score:", best_f1)
print("Best Threshold:", best_threshold)
```
