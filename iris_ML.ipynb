```python
import numpy as np

import pandas as pd

import seaborn as sns


file_path = 'C:/Users/7kingdom/Downloads/iris/iris.data'

column_names = ["sepal length in cm", "sepal width in cm", "petal length in cm", "petal width in cm", "class"]          

df = pd.read_csv(file_path, header=None, names=column_names)

df
```

```python
df = df.drop(columns=['class'])

df
```

```python
x = df.drop(columns=['petal width in cm'])

x
```

```python
y = df["petal width in cm"]

y
```

```python

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

```python
svm_regressor = SVR()
svm_regressor.fit(x_train, y_train)  

svm_predictions = svm_regressor.predict(x_test)

svm_mse = mean_squared_error(y_test, svm_predictions)
print("SVM MSE:", svm_mse)
```

```python
lr_regressor = LinearRegression()
lr_regressor.fit(x_train, y_train) 

lr_predictions = lr_regressor.predict(x_test)

lr_mse = mean_squared_error(y_test, lr_predictions)
print("Linear Regression MSE:", lr_mse)
```

```python
dt_regressor = DecisionTreeRegressor()
dt_regressor.fit(x_train, y_train)

dt_predictions = dt_regressor.predict(x_test)

dt_mse = mean_squared_error(y_test, dt_predictions)
print("Decision Tree MSE:", dt_mse)
```

```python
rf_regressor = RandomForestRegressor()
rf_regressor.fit(x_train, y_train)  

rf_predictions = rf_regressor.predict(x_test)

rf_mse = mean_squared_error(y_test, rf_predictions)
print("Random Forest MSE:", rf_mse)
```
