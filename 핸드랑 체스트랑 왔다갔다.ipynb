```python
from sklearn.model_selection import train_test_split

HX_train, HX_test, Hy_train, Hy_test = train_test_split(scaled_h_x, 
                                                    scaled_h_x["label"],
                                                    test_size = 0.95, 
                                                    shuffle = True, 
                                                    random_state = 21)
```

```python
from sklearn.model_selection import train_test_split

CX_train, CX_test, Cy_train, Cy_test = train_test_split(scaled_c_x, 
                                                    scaled_c_x["label"],
                                                    test_size = 0.95, 
                                                    shuffle = True, 
                                                    random_state = 21)
```

```python
TIME_STEPS = 64
STEP = 64
```

```python
from scipy import stats

def create_dataset(X, y, time_steps=1, step=1):
    Xs, ys = [], []
    for i in range(0, len(X) - time_steps, step):
        v = X.iloc[i:(i + time_steps)].values
        labels = y.iloc[i: i + time_steps]
        Xs.append(v)        
        ys.append(stats.mode(labels)[0][0])
    return np.array(Xs), np.array(ys).reshape(-1, 1)


HXs_train, Hys_train = create_dataset(HX_train[['handAcc16_1' , 'handAcc16_2' , 'handAcc16_3' ,
    'handAcc6_1' , 'handAcc6_2' , 'handAcc6_3' ,
    'handGyro1' , 'handGyro2',   'handGyro3' ,      
    'handMagne1',  'handMagne2' , 'handMagne3']], HX_train.label, TIME_STEPS, STEP)
HXs_test, Hys_test = create_dataset(HX_test[['handAcc16_1' , 'handAcc16_2' , 'handAcc16_3' ,
    'handAcc6_1' , 'handAcc6_2' , 'handAcc6_3' ,
    'handGyro1' , 'handGyro2',   'handGyro3' ,      
    'handMagne1',  'handMagne2' , 'handMagne3']], HX_test.label, TIME_STEPS, STEP)

print(HXs_train.shape, Hys_train.shape)
```

```python
CXs_train, Cys_train = create_dataset(CX_train[['chestAcc16_1', 'chestAcc16_2', 'chestAcc16_3', 'chestAcc6_1',
       'chestAcc6_2', 'chestAcc6_3', 'chestGyro1', 'chestGyro2', 'chestGyro3',
       'chestMagne1', 'chestMagne2', 'chestMagne3']], CX_train.label, TIME_STEPS, STEP)
CXs_test, Cys_test = create_dataset(CX_test[['chestAcc16_1', 'chestAcc16_2', 'chestAcc16_3', 'chestAcc6_1',
       'chestAcc6_2', 'chestAcc6_3', 'chestGyro1', 'chestGyro2', 'chestGyro3',
       'chestMagne1', 'chestMagne2', 'chestMagne3']], CX_test.label, TIME_STEPS, STEP)

print(CXs_train.shape, Cys_train.shape)
```

```python
from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder(handle_unknown = "ignore", sparse = False)
enc = enc.fit(Hys_train)

Hyss_train = enc.transform(Hys_train)
Hyss_test = enc.transform(Hys_test)
```

```python
from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder(handle_unknown = "ignore", sparse = False)
enc = enc.fit(Cys_train)

Cyss_train = enc.transform(Cys_train)
Cyss_test = enc.transform(Cys_test)
```

```python
print("HXs_train.shape: ", HXs_train.shape)
print("HXs_test.shape ", HXs_test.shape)
print("Hyss_train.shape ", Hyss_train.shape)
print("Hyss_test.shape ", Hyss_test.shape)
```

```python

verbose,epochs,batch_size=1,10,32 

n_timesteps,n_features,n_outputs=HXs_train.shape[1],HXs_train.shape[2],Hyss_train.shape[1]
```

```python
n_timesteps, n_features, n_outputs
# 40에 10
```

<h3>핸드로 모델 만들기<h3>

```python

inputs=keras.Input(shape=(n_timesteps,n_features))

conv_1=tf.keras.layers.Conv1D(filters=64,kernel_size=5,strides=2,activation='relu')(inputs)
maxpool_1=tf.keras.layers.MaxPooling1D(pool_size=2,strides=2)(conv_1)

conv_2=tf.keras.layers.Conv1D(filters=96,kernel_size=3,strides=1,activation='relu')(maxpool_1)
avg_pooling=tf.keras.layers.GlobalAveragePooling1D()(conv_2)
batch_norm=tf.keras.layers.BatchNormalization()(avg_pooling)

output=tf.keras.layers.Dense(n_outputs,activation='softmax')(batch_norm)
h_model=tf.keras.Model(inputs=inputs,outputs=output)
h_model.summary()
```

```python
h_model.compile(optimizer=Adam(learning_rate = 0.001), loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
history = h_model.fit(HXs_train, Hyss_train, epochs = 30, batch_size = 32, validation_split = 0.1, shuffle = True)
print(h_model.summary())
```

```python
from mlxtend.plotting import plot_confusion_matrix
from sklearn.metrics import confusion_matrix
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

<h3>헨드로 핸드<h3>

```python
base_loss,base_accuracy=h_model.evaluate(HXs_test,Hyss_test,batch_size=batch_size,verbose=1)
  
score=base_accuracy*100
print('Accuracy >{:f}'.format(score))
print('Base Loss >{:.2f}'.format(base_loss))
```

```python
y_test_arg=np.argmax(Hyss_test,axis=1)
Y_pred = np.argmax(h_model.predict(HXs_test), axis=1)

print('Confusion Matrix')
print(confusion_matrix(y_test_arg, Y_pred))
```
